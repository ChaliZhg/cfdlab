```python
#  p6 - variable coefficient wave equation
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.collections import LineCollection
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Set up grid and differentiation matrix:
N = 128; h = 2*np.pi/N; x = h*np.linspace(1,N,N);
t = 0.0; dt = h/4.0
c = 0.2 + np.sin(x-1.0)**2.0
v = np.exp(-100.0*(x-1.0)**2.0); vold = np.exp(-100.0*(x-0.2*dt-1.0)**2.0);

# Time-stepping by leap-frog formula
tmax = 8.0; tplot = 0.15;
plotgap = np.round(tplot/dt).astype(int); dt = tplot/plotgap;
nplots = np.round(tmax/tplot).astype(int)
data = []
data.append(list(zip(x, v)))
tdata = np.zeros(nplots+1)
for i in range(1,nplots):
    for n in range(plotgap):
        t = t + dt
        v_hat = np.fft.fft(v)
        w_hat = 1j*np.zeros(N)
        w_hat[0:N/2] = 1j*np.arange(0,N/2)
        w_hat[N/2+1:] = 1j*np.arange(-N/2+1,0,1)
        w_hat = w_hat * v_hat
        w = np.real(np.fft.ifft(w_hat))
        vnew = vold - 2*dt*c*w
        vold = v; v = vnew;
    data.append(list(zip(x, v)))
    tdata[i] = t;

plt.figsize(12,8)
fig = plt.figure()
ax = fig.gca(projection='3d')
poly = LineCollection(data)
poly.set_alpha(0.5)
ax.add_collection3d(poly, zs=tdata, zdir='y')
ax.set_xlabel('X')
ax.set_xlim3d(0, 2*np.pi)
ax.set_ylabel('Y')
ax.set_ylim3d(0, 8)
ax.set_zlabel('Z')
ax.set_zlim3d(0, 4)
ax.view_init(75,-85)
```

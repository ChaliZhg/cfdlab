```python
#  p5 - repetition of p4 via FFT
#        For complex v, delete "real" commands.
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Set up grid and differentiation matrix:
N = 24; h = 2*np.pi/N; x = h*np.linspace(1,N,N);

# Differentiation of a hat function:
v = np.maximum(0,1-np.abs(x-np.pi)/2)
v_hat = np.fft.fft(v)
w_hat = 1j*np.zeros(N)
w_hat[0:N/2] = 1j*np.arange(0,N/2)
w_hat[N/2+1:] = 1j*np.arange(-N/2+1,0,1)
w_hat = w_hat * v_hat
w = np.real(np.fft.ifft(w_hat))

plt.figsize(10,6)

plt.subplot(3,2,1)
plt.plot(x,v,'.-')
plt.axis([0, 2*np.pi, -.5, 1.5])
plt.title('function')
plt.subplot(3,2,2)
plt.plot(x,w,'.-')
plt.axis([0, 2*np.pi, -1, 1])
plt.title('spectral derivative')

# Differentiation of exp(sin(x)):
v = np.exp(np.sin(x)); vprime = np.cos(x)*v;
v_hat = np.fft.fft(v)
w_hat = 1j*np.zeros(N)
w_hat[0:N/2] = 1j*np.arange(0,N/2)
w_hat[N/2+1:] = 1j*np.arange(-N/2+1,0,1)
w_hat = w_hat * v_hat
w = np.real(np.fft.ifft(w_hat))
plt.subplot(3,2,3)
plt.plot(x,v,'.-')
plt.axis([0, 2*np.pi, 0, 3])
plt.subplot(3,2,4)
plt.plot(x,w,'.-')
plt.axis([0, 2*np.pi, -2, 2])
error = np.linalg.norm(w-vprime,inf)
plt.text(2.0,1.4,"max error="+str(error))
```

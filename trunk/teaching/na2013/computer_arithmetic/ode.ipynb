```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def ode(alpha,h,N):
    y = np.zeros(N)
    t = np.zeros(N)
    t[0] = 0; y[0]=1
    t[1] = h; y[1]=np.exp(alpha*h)
    a = 2.0*alpha*h
    for i in range(1,N-1):
        y[i+1] = a*y[i] + y[i-1]
        t[i+1] = t[i] + h
    ye = np.exp(alpha*t)
    plt.plot(t,y,'o-',t,ye,'r-')
    plt.xlabel('t')
    plt.ylabel('y')
```

```python
alpha=1
h=0.1
N = 20
ode(alpha,h,N)
```

```python
alpha=-1
h=0.1
N = 100
ode(alpha,h,N)
```

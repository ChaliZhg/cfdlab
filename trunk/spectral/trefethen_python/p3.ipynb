```python
# p3 - band-limited interpolation
import numpy as np
from matplotlib import pyplot as plt
```

```python
h = 1.0;
xmax = 10.0;
x = np.arange(-xmax,xmax+h,h)
xx = np.arange(-xmax-h/20, xmax+h/20, h/10)
plt.figsize(5,10)
for pl in range(3):
    plt.subplot(4,1,pl+1)
    if pl==0:
        v = (x==0)                      # delta function
    elif pl==1:
        v = (np.abs(x) <= 3.0)          # square wave
    else:
        v = np.maximum(0,1-np.abs(x)/3) # hat function
    plt.plot(x,v,'.')
    plt.grid(True)
    p = np.zeros(len(xx))
    for i in range(len(x)):
        p = p + v[i]*np.sin(np.pi*(xx-x[i])/h)/(np.pi*(xx-x[i])/h)
    plt.plot(xx,p)
    plt.axis([-xmax,xmax,-0.5,1.5])
```

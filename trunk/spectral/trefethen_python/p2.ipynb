```python
from scipy.linalg import toeplitz
import numpy as np
from matplotlib import pyplot as plt
```

```python
plt.figsize(5,5)
for N in range(2,100,2):
    h = 2.0*np.pi/N
    x = -np.pi + np.arange(1,N+1)*h
    u = np.exp(np.sin(x))
    uprime = np.cos(x)*u
    col = np.zeros(N)
    col[1:] = 0.5*(-1.0)**np.arange(1,N)/np.tan(np.arange(1,N)*h/2.0)
    row = np.zeros(N)
    row[0] = col[0]
    row[1:] = col[N-1:0:-1]
    D = toeplitz(col,row)
    error = np.linalg.norm(D.dot(u)-uprime,inf)
    plt.loglog(N,error,'or')
    plt.hold(True)
    
plt.grid(True)
plt.xlabel('N')
plt.ylabel('error')
```

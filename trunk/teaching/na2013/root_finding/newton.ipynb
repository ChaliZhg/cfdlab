```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
```

```python
x=sym.Symbol('x')
# Define the function
f=sym.exp(x)-3.0/2.0-sym.atan(x)
# Compute its derivative
df=sym.diff(f,x)
# Make functions
F = sym.lambdify(x, f, modules=['numpy'])
DF = sym.lambdify(x, df, modules=['numpy'])
X = linspace(-1,1,100)
plt.subplot(1,2,1)
plt.plot(X,F(X))
plt.grid('on')
plt.title('Function')
plt.subplot(1,2,2)
plt.plot(X,DF(X))
plt.grid('on')
plt.title('Derivative')
```

```python
M = 20      # maximum iterations
x = 0.5     # initial guess
eps = 1e-16 # relative tolerance on root

f = F(x)
for i in range(M):
    df= DF(x)
    dx= -f/df
    x = x + dx
    e = np.abs(dx)
    f  = F(x)
    print "{0:6d} {1:22.14e} {2:22.14e} {3:22.14e}".format(i,x,e,np.abs(f))
    if e < eps * np.abs(x):
        break
```
